```python
import os
import shutil
import random
import os

data_path = '../data/src'
class_names = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]


for item in class_names:
    data_dir = f'../data/src/{item}'  # Thay thế bằng đường dẫn thực tế
    # Thư mục để lưu dữ liệu train và test
    train_dir = f'../data/train/{item}'
    test_dir = f'../data/test/{item}'
    # Tạo thư mục train và test nếu chưa tồn tại
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # Lấy danh sách tất cả các tệp trong thư mục dữ liệu gốc
    files = os.listdir(data_dir)

    # Lọc ra chỉ các tệp ảnh .jpg
    files = [file for file in files if file.endswith('.jpg')]

    # Xáo trộn danh sách tệp
    random.shuffle(files)

    # Tính toán số lượng tệp cho tập train và test
    split_index = int(len(files) * 0.8)  # 80% cho train

    # Chia dữ liệu vào thư mục train và test
    train_files = files[:split_index]
    test_files = files[split_index:]

    for file in train_files:
        shutil.copy(os.path.join(data_dir, file), train_dir)

    for file in test_files:
        shutil.copy(os.path.join(data_dir, file), test_dir)

    print("Dữ liệu đã được chia thành công vào thư mục train và test.")
```

```python
#CODE TEST
import numpy as np
import os, uuid
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
# Dự đoán với mô hình
def predict_image(img_path):
    model = load_model('../cnn_model_improved.keras')
    img = image.load_img(img_path, target_size=(150, 150))  # Thay đổi kích thước hình ảnh
    img_array = image.img_to_array(img) / 255.0  # Chia pixel cho 255
    img_array = np.expand_dims(img_array, axis=0)  # Thêm chiều cho batch

    # Dự đoán
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions, axis=1)
    return predicted_class

# Gọi hàm dự đoán với một hình ảnh cụ thể
rs = predict_image('../data/image/5a5f4c26-832c-4eb9-a9d4-8817f2f29a84.jpg')  # Thay đổi đường dẫn đến hình ảnh của bạn
data_path = '../data/test'
class_names = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]
print(f'result: {class_names[int(rs)]}')
```

```python
# code sinh anh
import cv2
import os
import numpy as np

def generate_images(input_path, output_path, n):
    # Đọc ảnh
    image = cv2.imread(input_path)

    # Kiểm tra xem ảnh có được đọc thành công không
    if image is None:
        print("Không thể đọc ảnh. Vui lòng kiểm tra đường dẫn.")
        return

    # Danh sách chứa các biến thể của ảnh
    transformations = [
        ('original', image),
        ('flip_horizontal', cv2.flip(image, 1)),  # Lật ngang
        ('flip_vertical', cv2.flip(image, 0)),    # Lật dọc
    ]

    # Xoay ảnh động từ -30 đến 30 độ
    for angle in range(-80, 81, 0.5):  # Thay đổi bước nhảy tại đây (ví dụ: 10 độ)
        # Tính toán kích thước của ảnh sau khi xoay
        (h, w) = image.shape[:2]
        center = (w // 2, h // 2)
        M = cv2.getRotationMatrix2D(center, angle, 1.0)  # Tạo ma trận xoay
        rotated_image = cv2.warpAffine(image, M, (w, h))  # Áp dụng ma trận xoay
        transformations.append((f'rotate_{angle}', rotated_image))

    # Tạo các ảnh biến thể và lưu chúng
    count = 0
    for name, img in transformations:
        if count < n:  # Kiểm tra nếu chưa đủ n ảnh
            output_file = os.path.join(output_path, f'{name}.jpg')
            cv2.imwrite(output_file, img)
            count += 1
        else:
            break

    print(f'Tạo thành công {count} ảnh.')

# Đường dẫn ảnh gốc
input_image_path = 'data/anh.jpg'
# Đường dẫn thư mục lưu ảnh biến thể
output_directory = 'data/output'

# Kiểm tra xem thư mục đầu ra có tồn tại không, nếu không thì tạo mới
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Số lượng ảnh muốn tạo

# Tạo ảnh biến thể
generate_images(input_image_path, output_directory, 300)
```
